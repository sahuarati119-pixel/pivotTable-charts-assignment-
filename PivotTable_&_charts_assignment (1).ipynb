3. Apply Conditional Formatting to highlight sales greater than 400


*   Select the Sales column
*   Go to Home → Conditional Formatting → Highlight Cells Rules
*   Select "Greater Than"
*   Enter value 400
*   Choose a color (like red or green)

 -Now, sales greater than 400 will be highlighted automatically.
